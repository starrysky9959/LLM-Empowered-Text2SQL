```python
import json
import re

from numpy import extract

pattern = re.compile(r'"question": "(.*)",\s*"sql": "(.*)"')
err = []
for i in range(1, 8659 + 1):
    with open(f"oss_insight/new_dataset_{i}.json", "r") as input_file:
        output = []
        text = input_file.read()
        matches = re.findall(pattern, text)
        if len(matches) != 10:
            err.append(i)
            continue

        for match in matches:
            question = match[0]
            sql = match[1]
            output.append({"question": question, "sql": sql})
            # print("Question:", question)
            # print("SQL:", sql)
        with open(f"extract/{i}.json", "w") as output_file:
            json.dump(output, output_file)
print("err",err)
```

```python
pattern = re.compile(r'"question": "([^"]*)",\s*"sql": "([^"]*)"', re.MULTILINE)
new_err=[]
for i in err:
    with open(f"oss_insight/new_dataset_{i}.json", "r") as input_file:
        output = []
        text = input_file.read()
        print("i:", i)
        matches = re.findall(pattern, text)
        print(len(matches))
        if len(matches) != 10:
            new_err.append(i)
            continue

        for match in matches:
            question = match[0]
            sql = match[1]
            output.append({"question": question, "sql": sql})
            # print("Question:", question)
            # print("SQL:", sql)
        # break
        with open(f"extract/{i}.json", "w") as output_file:
            json.dump(output, output_file)
print(new_err)
```
